```python
import json
import pandas as pd
```

```python
df00 = pd.read_csv("Data/final_tmdb_data_2000.csv.gz", low_memory = False)
df00.head()
```

```python
df00['imdb_id']
```

```python
df01 = pd.read_csv("Data/final_tmdb_data_2001.csv.gz", low_memory = False)
df01.head()
```

```python
df01['imdb_id']
```

```python
movies = pd.concat([df00,df01]) 
movies.head()
```

```python
movies.info()
```

How many movies have at least some valid financial information (> 0 for budget OR revenue)?

```python
finac = (movies['budget'] > 0) | (movies['revenue'] > 0)
finac.value_counts()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure(figsize = (10,8))
sns.countplot(x = finac)
plt.title('Count of some valid financial information');
```

```python
a_budget = movies['budget'] > 0
some_revenue = movies['revenue'] > 0
fin_df = movies.loc[a_budget | some_revenue,:]
fin_df.head()
```

How many movies are there in each of the certification categories?

```python
fin_df.groupby('certification')['imdb_id'].count()
```

```python
plt.figure(figsize = (10,8))
sns.countplot(x = 'certification', data = fin_df)
plt.title('Movies in each certification category');
```

What is the average revenue per certification category?

```python
fin_df.groupby('certification')['revenue'].mean()
```

```python
plt.figure(figsize = (10,8))
sns.barplot(x = 'certification', y = 'revenue', data = fin_df, ci = None)
plt.title('Average revenue for each certification category');
```

What is the average budget per certification category?

```python
fin_df.groupby('certification')['budget'].mean()
```

```python
plt.figure(figsize = (10,8))
sns.barplot(x = 'certification', y = 'budget', data = fin_df, ci = None)
plt.title('Average budget for each certification category');
```
